## <blockquote><b> Data Dictionnary </b></blockquote>

<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true"

Le jeu de données utilisé pour notre étude est structuré de la façon suivante :

|**Variable**       |**Description**                                                                                   |
|---------------------|---------------------------------------------------------------------------------------------------|
| `step`             | Représente une unité de temps dans le monde réel. 1 pas correspond à 1 heure. Total : 744 étapes (30 jours). |
| `type`             | Type de transaction : ENTRÉE, SORTIE, DÉBIT, PAIEMENT et TRANSFERT.                               |
| `montant`          | Montant de la transaction en monnaie locale.                                                     |
| `nameOrig`         | Client à l'origine de la transaction.                                                            |
| `oldbalanceOrg`    | Solde initial du client à l'origine avant la transaction.                                        |
| `newbalanceOrig`   | Nouveau solde du client à l'origine après la transaction.                                        |
| `nameDest`         | Client qui est le destinataire de la transaction.                                                |
| `oldbalanceDest`   | Solde initial du destinataire avant la transaction. Note : Pas d'informations pour les clients commençant par "M" (Merchants). |
| `newbalanceDest`   | Nouveau solde du destinataire après la transaction. Note : Pas d'informations pour les clients commençant par "M" (Merchants). |
| `isFraud`          | Indicateur binaire (1 ou 0) indiquant si la transaction est frauduleuse. Ces transactions sont effectuées par des agents frauduleux dans le cadre de la simulation. |
|`isFlaggedFraud`     | Variable binaire (1 ou 0), indique les tentatives illégales (suspectes) détectées|

<a id='ind'></a>

## <blockquote><b> Indicateurs retenus </b> </blockquote>

<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true"
style="color:#007DD2;font-size:85%; " base-toggle="popover"><b>Table des matières^</b></a>

> 1. **Nombre de Transactions**  
- **Indicateur** : Nombre total de transactions dans le jeu de données.  
- **Utilité** : Donne une vue d'ensemble de l'activité financière.  

> 2. **Répartition des Types de transactions**  
- **Indicateur** : Pourcentage ou nombre de transactions par type (CASH-IN, CASH-OUT, DEBIT, PAYMENT, TRANSFER).  
- **Utilité** : Identifie les types de transactions les plus fréquents.  

> 3. **Nombre et taux de transactions Frauduleuses (isFraud)**
- **Indicateur** : Pourcentage de transactions frauduleuses par rapport au total des transactions.  
- **Utilité** : Mesure la proportion de fraude dans l'ensemble des transactions.  

> 4. **Nombre et taux de transactions Suspectes (isFlaggedFraud)**  
- **Indicateur** : Nombre de transactions marquées comme suspectes (isFlaggedFraud = 1).  
- **Utilité** : Identifie les transactions qui ont déclenché des alertes de fraude.  

> 5. **Montant des transactions Frauduleuses**  
- **Indicateur** : Somme des montants des transactions frauduleuses.  
- **Utilité** : Évalue l'impact financier de la fraude.  

> 6. **Fréquence des Transactions par Client (nameOrig)**  
- **Indicateur** : Nombre de transactions initiées par chaque client.  
- **Utilité** : Détecte les clients avec une activité anormalement élevée.  

> 7. **Fréquence des Transactions par Destinataire (nameDest)**  
- **Indicateur** : Nombre de transactions reçues par chaque destinataire.  
- **Utilité** : Identifie les destinataires avec un volume de transactions suspect.  

> 8. **Soldes des Clients**  
- **Indicateur** : Différence entre newbalanceOrig et oldbalanceOrg pour chaque transaction.  
- **Utilité** : Identifie les clients dont les soldes changent de manière suspecte.  

> 9. **Transactions avec un Solde Initial Nul (oldbalanceOrg = 0)**  
- **Indicateur** : Nombre de transactions où le solde initial de l'initiateur est nul.  
- **Utilité** : Peut indiquer des comptes nouvellement créés ou des comportements suspects.  

> 10. **Transactions avec un Solde Final Nul (newbalanceOrig = 0)**  
- Indicateur : Nombre de transactions où le solde final de l'initiateur est nul.  
- **Utilité** : Peut indiquer des tentatives de vider un compte.  

> 11. **Transactions pour lesquelles *nameDest* commence par *'M'***
- **Indicateur** : Types particuliers d'individus dans le jeu de données
- **Utilité** : Peut permettre d'identifier un certains nombres de destinataires

**Recherche de valeurs manquantes**

```python
cols_finance = finance_data.columns


def compte_valeurs_manquante(col_name):
  '''Cette fonction permet de compter le nombre de valeurs manquantes pour une variable donnee'''
  return finance_data.filter(col(col_name).isNull()).count()

###avoir l effectif des valeurs manquantes par variables
missing_counts = {col_name: compte_valeurs_manquante(col_name) for col_name in cols_finance}
print(missing_counts)

## afficher le tout dans un dataframe
missing_info = spark.createDataFrame(
    [Row(Variable=col_name, Missing_Count=missing_counts[col_name])
     for col_name in cols_finance]
)
missing_info.show()
```

On note qu'aucune des variable n'a pas de valeurs manquantes. Ce qui est très interressant car nous n'aurront pas à faire des imputations qui biaisent la qualité d'informations contenue dans les données pour pouvoir faire nos analyses

```python
def type_variables(df = finance_data):
  '''Fonction Generaliste pour explorer le types de variables dans un  dataframe quelcquonque'''
  schema_info = [(field.name, str(field.dataType)) for field in df.schema]
  schema_df = spark.createDataFrame(schema_info, ["Variables", "Type"])
  return schema_df

type_variables().show()
```

On remarque que par defaut il importe tout sous forme de chaine de charactere il faudra donc redefinir les types de chaque colonne

### <b> *Rédéfinition des types des variables* <b>

<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true"

```python
finance_data = finance_data.withColumn("step",col("step").cast(IntegerType()))
finance_data = finance_data.withColumn("amount",col("amount").cast(FloatType()))
finance_data = finance_data.withColumn("oldbalanceOrg",col("oldbalanceOrg").cast(FloatType()))
finance_data = finance_data.withColumn("newbalanceOrig",col("newbalanceOrig").cast(FloatType()))
finance_data = finance_data.withColumn("oldbalanceDest",col("oldbalanceDest").cast(FloatType()))
finance_data = finance_data.withColumn("newbalanceDest",col("newbalanceDest").cast(FloatType()))
finance_data = finance_data.withColumn("isFraud",col("isFraud").cast(IntegerType()))
finance_data = finance_data.withColumn("isFlaggedFraud",col("isFlaggedFraud").cast(IntegerType()))
finance_data = finance_data.withColumn("type",col("type").cast(StringType()))
finance_data = finance_data.withColumn("nameDest",col("nameDest").cast(StringType()))
finance_data = finance_data.withColumn("nameOrig",col("nameOrig").cast(StringType()))
```

```python
type_variables(finance_data).show()
```

```python
def compte_per_modal(variable="step"):
  '''faire un comptage par l heure de simulation du moi'''

  return finance_data.groupBy(variable).agg(count("*").alias("count"))

compte_per_modal().show()
```

On remarque une distribution non uniforme des trafics en heures de transactions. Il serait donc pertinent de se ramener au jour, a la semaine du mois de simulation afin de faire des analyses plus pousse

<a id='conversion'></a>

### <b> Conversion des heures de simulation en jours du mois sur la durée des simulations <b>

<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true"
style="color:#007DD2;font-size:85%; " base-toggle="popover"><b>Table des matières^</b></a>

```python
def conversion_en_jour_du_mois(heure):
  """Cette fonction permet de convertir les heures(step) en jour """

  return heure // 24 +1


conversion_en_jour_du_mois_udf = udf(conversion_en_jour_du_mois, IntegerType()) # ici on precise la fonction et le type de la fonction

finance_data = finance_data.withColumn("jour_du_mois", conversion_en_jour_du_mois_udf(col("step")))
        # Ensuite on l utilise pour creer une nouvelle colonne
```

```python
compte_per_modal('jour_du_mois').orderBy(col('count').desc()).show()
```

_Donne les pourcentage ou nombre de transactions par type de transactions `(CASH-IN, CASH-OUT, DEBIT, PAYMENT, TRANSFER)`_

<a id='c2.1'></a>

## **| Calcul de la répartition à l'aide du concept de MapReduce**

<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true"
style="color:#007DD2;font-size:85%; " base-toggle="popover"><b>Table des matières^</b></a>

```python
map = finance_data.select('type')\
        .rdd\
        .map(lambda row: (row['type'], 1))  ##transformation
reduce_ = map.reduceByKey(lambda a, b: a + b)                         ## action : grouper suivant une cles

df = reduce_.toDF().select(col('_1').alias('type'),col('_2').alias('Effectif'))

eff_type = df.toPandas()  ## transformation en data frame
df.show()
```

<a id='c2.2'></a>

## **| Répartition selon les Types de transactions**

<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true"
style="color:#007DD2;font-size:80%; " base-toggle="popover"><b>Table des matières^</b></a>

```python
eff_type.sort_values(by='Effectif',ascending=False).set_index('type').style.bar(subset=['Effectif'], color='#13C5E3')
```

```python
fig = px.bar(eff_type.sort_values(by='Effectif'),
             y='type', x='Effectif',
             title='Répartition des types de transactions',text_auto='.0f')
fig.update_layout(xaxis=dict(title=""))
fig.show()
```

<a id='c2.3'></a>

## **| Suivi de l'indicateur par jour**

<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true"
style="color:#007DD2;font-size:80%; " base-toggle="popover"><b>Table des matières^</b></a>

```python
df = finance_data.groupBy('type','jour_du_mois').agg(count("*").alias("count"))\
            .orderBy(col('type').asc(),col('jour_du_mois').asc())

eff_type_jour = df.toPandas()
```

```python
fig = px.line(eff_type_jour, x='jour_du_mois', y='count', color='type', title='Répartition des types de transactions par jour')
fig.update_layout(
    yaxis=dict(title="Effectif"),
    legend=dict(
    orientation="h",
    yanchor="bottom",
    y=-0.2,
    xanchor="center",
    x=0.5
))

fig.show()
```

## **| Nombre total de transactions avec un solde initial nul**

## **| Calcul par MapReduce**

<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true"
style="color:#007DD2;font-size:80%; " base-toggle="popover"><b>Table des matières^</b></a>

```python
# Par un filtrage simple
filtered_data = finance_data.rdd.filter(lambda row: row['nameDest'].startswith('M'))
filtered_df = filtered_data.toDF()
nb_ = filtered_df.count()

print(f"Le nombre total de transactions dont nameDest commence par 'M' est {nb_:,}")
print()

filtered_df.show()
```

```python
# on remplace tout ce qui ne commance pas par M par none ou zero
mapped_rdd = finance_data.rdd.map(lambda row: (row['nameDest'], 1) if row['nameDest'].startswith('M') else (None, 0))

# Reduce : faire le comptage par cle.
reduced_rdd = mapped_rdd.reduceByKey(lambda x, y: x + y)

# On enleve les Null ou 0
filtered_rdd = reduced_rdd.filter(lambda x: x[0] is not None)

result_df = filtered_rdd.toDF(['nameDest', 'Transaction'])
result_df.orderBy(col('Transaction').desc()).show(10)
```

<a id='c11.2'></a>

## **| Nombre de transactions par jour où _nameDest_ commence par *M***

<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true"
style="color:#007DD2;font-size:80%; " base-toggle="popover"><b>Table des matières^</b></a>

```python
# Même idées que precedemment
mapped_rdd = finance_data.rdd.map(lambda row: (row['jour_du_mois'], 1) if row['nameDest'].startswith('M') else (row['jour_du_mois'], 0))

# sommation
reduced_rdd = mapped_rdd.reduceByKey(lambda x, y: x + y)

result_df = reduced_rdd.toDF(['jour_du_mois', 'TransactionCount'])
result_df.show()
```

```python
result_pd_df = result_df.toPandas().sort_values(by='jour_du_mois')

fig = px.line(result_pd_df,
              x='jour_du_mois', y='TransactionCount',
              markers=True,
              title='Nombre de transactions par jour où nameDest commence par M')
fig.update_layout(yaxis=dict(title="Nb de transactions"))
fig.show()
```
